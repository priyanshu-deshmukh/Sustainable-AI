```python
import numpy as np
```

```python
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
```

```python
#remove missing values using dropna

cleaned_data = {k: v for k, v in data.items() if v is not None}
print(data)
```

```python
#replace missing value by mean using fillna

cleaned_data = {k: v if v is not np.nan else np.mean(v) for k, v in cleaned_data.items()}
print(cleaned_data)
```

```python
#code for forward fill missing values

cleaned_data = {k: v if v is not np.nan else v[1] for k, v in cleaned_data.items()}
print(cleaned_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = {
    "Energy Source": ["Solar", "Wind", "Coal", "Natural Gas", "Nuclear", "Hydro"],
    "Energy Consumption (MWh)": [1500, 1200, 3000, 2500, 1800, 1000]
}

# Create DataFrame
cleaned_data = pd.DataFrame(data)

# Plotting
labels = cleaned_data["Energy Source"]
sizes = cleaned_data["Energy Consumption (MWh)"]
explode = (0, 0.1, 0, 0, 0, 0)

fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

ax1.axis('equal')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = {
    "Energy Source": ["Solar", "Wind", "Coal", "Natural Gas", "Nuclear", "Hydro"],
    "Energy Consumption (MWh)": [1500, 1200, 3000, 2500, 1800, 1000]
}

# Create DataFrame
cleaned_data = pd.DataFrame(data)

# Generate random carbon emissions
np.random.seed(42)  # For reproducibility
cleaned_data["Carbon Emissions (tons)"] = np.random.randint(100, 1000, len(cleaned_data))

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(cleaned_data["Energy Consumption (MWh)"], 
            cleaned_data["Carbon Emissions (tons)"], 
            color='skyblue')

plt.xlabel("Energy Consumption (MWh)")
plt.ylabel("Carbon Emissions (tons)")
plt.title("Energy Consumption vs Carbon Emissions")
plt.grid(True)

# Annotate points
for i, txt in enumerate(cleaned_data["Energy Source"]):
    plt.annotate(txt, (cleaned_data["Energy Consumption (MWh)"][i], 
                       cleaned_data["Carbon Emissions (tons)"][i]))

plt.show()

#checking the linearity
```
